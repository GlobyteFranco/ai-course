# Prueba simple de Tools - Mundiales

Notebook minimo para probar las tools definidas en `src/tools/tools.py`.

```python
from pathlib import Path
import sys

# Resolver raiz del proyecto para poder importar src.tools.tools
BASE_DIR = Path.cwd().resolve()
for candidate in [BASE_DIR, *BASE_DIR.parents]:
    if (candidate / "src").exists() and (candidate / "data").exists():
        BASE_DIR = candidate
        break

if str(BASE_DIR) not in sys.path:
    sys.path.append(str(BASE_DIR))

from src.tools.tools import (
    matches_by_year,
    top_attendance,
    avg_goals_by_stage,
    get_worldcup_tools,
)

print("BASE_DIR:", BASE_DIR)
```

```python
# Pruebas directas de funciones
print(matches_by_year(2014))
print("\n" + "-" * 70 + "\n")
print(top_attendance(3))
print("\n" + "-" * 70 + "\n")
print(avg_goals_by_stage("Final"))
```

```python
# Prueba de lista de tools (para LangChain / LangGraph)
tools = get_worldcup_tools()

nombres = [getattr(t, "name", getattr(t, "__name__", "tool")) for t in tools]
print("Tools disponibles:", nombres)

# Si no hay decorador @tool, igual funcionan como funciones Python
if callable(tools[0]) and not hasattr(tools[0], "invoke"):
    print("\nInvocacion directa:")
    print(tools[0](1930))
else:
    print("\nInvocacion estilo LangChain:")
    print(tools[0].invoke({"year": 1930}))
```
