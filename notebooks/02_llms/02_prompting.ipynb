```python
plantilla_colombiana = """Responde a la pregunta con base en el siguiente contexto, parce. Si la
pregunta no puede ser respondida con la información proporcionada, responde
con "No tengo ni idea, ome".

Contexto: {contexto}
Pregunta: {pregunta}

Respuesta (escribe como colombiano informal): """

prompt_plantilla_colombiana = PromptTemplate(
    input_variables=["pregunta", "contexto"],
    template=plantilla_colombiana
)
```

```python
# llm_chain = LLMChain(
#     prompt=prompt_plantilla_colombiana,
#     llm=llm
# )

modern_chain = prompt_plantilla_colombiana | llm

pregunta = "Que nos quiere contar la autora J.K Rowling con su ultima obra?"
contexto = "Sos el director de la UTN, sabes mucho de ingeniera pero del resto de especialidades no. Aun asi por orgullo vas a hablar con total seguridad inclusive aunque sepas poco del tema de la pregunta"

# llm.invoke(pregunta)
# Así se llamaría la chain clásica (pero va a fallar)
respuesta = modern_chain.invoke({"pregunta":pregunta, "contexto":contexto})
print(respuesta.content)
```
