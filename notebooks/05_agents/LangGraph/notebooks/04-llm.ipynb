```python
# Basic libraries
import os
from dotenv import load_dotenv

# LLMs
from langchain_openai import ChatOpenAI
from langchain_ollama import ChatOllama
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

load_dotenv()
```

# LLM fallback: OpenAI -> Gemini -> Ollama

```python
try:
    # Intentar OpenAI primero
    gtp_llm = ChatOpenAI(
        model="gpt-4o-mini",
        temperature=0,
        api_key=os.getenv("OPENAI_API_KEY"),
    )
    _ = gtp_llm.invoke("ping")
    provider = "OpenAI"
except Exception:
    # Si OpenAI falla, usar Ollama
    gtp_llm = ChatOllama(
        model=os.getenv("OLLAMA_MODEL", "llama3.2:3b"),
        temperature=0,
    )
    _ = gtp_llm.invoke("ping")
    provider = "Ollama"

print(f"Proveedor activo: {provider}")
response = gtp_llm.invoke("Hello, how are you?")
response.text
```

```python
messages_1 = SystemMessage(content="You are a helpful assist that talks in spanish")
messages_2 = HumanMessage(content="ME llamo Eder")
messages_3 = AIMessage(content="Hola Eder, ¿cómo estás?")
messages_4 = HumanMessage(content="Cómo me llamo?")
history = [messages_1, messages_2, messages_3, messages_4]
response = gtp_llm.invoke(history)

response.pretty_print()
```
